```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
titanic = sns.load_dataset('titanic')  
print("üîç Preview of dataset:")
print(titanic.head())
print("\nüßπ Cleaning data...")
titanic.drop(columns=['deck', 'embark_town'], inplace=True)
titanic['age'].fillna(titanic['age'].median(), inplace=True)
titanic['embarked'].fillna(titanic['embarked'].mode()[0], inplace=True)
titanic.dropna(inplace=True)
print("\nüìä Performing EDA...")
print("\nüìà Summary statistics:")
print(titanic.describe(include='all'))
sns.countplot(data=titanic, x='survived')
plt.title("Survival Count")
plt.xlabel("Survived (0 = No, 1 = Yes)")
plt.ylabel("Count")
plt.show()
sns.countplot(data=titanic, x='sex', hue='survived')
plt.title("Survival by Sex")
plt.xlabel("Sex")
plt.ylabel("Count")
plt.show()
sns.histplot(data=titanic, x='age', hue='survived', kde=True, bins=30)
plt.title("Age Distribution by Survival")
plt.xlabel("Age")
plt.ylabel("Density")
plt.show()
sns.countplot(data=titanic, x='class', hue='survived')
plt.title("Survival by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Count")
plt.show()
plt.figure(figsize=(10,6))
sns.heatmap(titanic.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
sns.pairplot(titanic[['age', 'fare', 'pclass', 'survived']], hue='survived')
plt.suptitle("Pairplot of Numerical Features", y=1.02)
plt.show()
print("\nüìå Insights & Trends:")
print("""
1. Women had a significantly higher survival rate than men.
2. Passengers in 1st class had higher survival rates than those in 2nd and 3rd class.
3. Younger passengers had slightly higher chances of survival.
4. Higher fare passengers (often 1st class) were more likely to survive.
5. There is a moderate correlation between Fare and Pclass (lower class = cheaper ticket).
""")
```
